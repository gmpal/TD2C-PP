# Results: Zero-Shot Real-World Transfer

The ultimate test is applying the model—trained *only* on synthetic mathematical functions—to realistic biological data.

We evaluated TD2C on:
1.  **NetSim:** Simulated fMRI data (Brain networks).
2.  **DREAM3:** Gene regulatory networks (E. coli and Yeast).

### The Challenge
TD2C was never shown biological data during training. It must generalize based solely on the fundamental properties of information flow.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Placeholder data representing the paper's results 
# (Use this if the user hasn't run the full benchmark yet)
data = {
    'Dataset': ['NetSim-5', 'NetSim-5', 'NetSim-10', 'NetSim-10', 'DREAM3-10', 'DREAM3-10'],
    'Method': ['TD2C', 'PCMCI', 'TD2C', 'PCMCI', 'TD2C', 'PCMCI'],
    'F1-Score': [0.52, 0.42, 0.49, 0.41, 0.36, 0.35] # Values from paper Table
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Dataset', y='F1-Score', hue='Method', palette=['#CC4125', '#8DA0CB'])
plt.title("Zero-Shot Performance on Realistic Benchmarks")
plt.ylabel("F1 Score (Higher is better)")
plt.ylim(0, 0.6)
plt.grid(axis='y', alpha=0.3)
plt.show()
```

### Analysis
TD2C consistently outperforms or matches the state-of-the-art constraint-based method (PCMCI). Crucially, TD2C achieves higher **Recall** (finding more true gene interactions) without a significant drop in precision.
