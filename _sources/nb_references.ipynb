# Appendix A: List of Causal Descriptors

TD2C computes a 63-dimensional feature vector for every candidate edge $X \to Y$. The features are categorized into four families.

### 1. Information Theoretic (28 features)
Based on Mutual Information (MI) and Conditional MI (CMI), estimated via the KSG non-parametric estimator.
*   `transfer_entropy_fwd`: $I(X_{t-1}; Y_t \mid Y_{t-1})$
*   `te_asymmetry`: Average $\Delta TE$ over 15 lags.
*   `eff_cau`: Symmetric MI $I(X; Y)$.
*   ...

### 2. Error-Based (2 features)
Based on residuals from Ridge regression.
*   `parcorr_errors`: Partial correlation of residuals.
*   `resid_corr`: Correlation of $Y$'s residuals with $X$.

### 3. Statistical Moments (12 features)
Capturing non-Gaussianity.
*   `skewness_ca`, `skewness_ef`
*   `kurtosis_ca`, `kurtosis_ef`
*   `HOC_3_1`: Higher order cross-moment $E[X^3 Y^1]$.

### 4. Linear Coefficients (2 features)
*   `coeff_cause`: Regression coefficient of $X$ predicting $Y$.
*   `coeff_eff`: Regression coefficient of $Y$ predicting $X$.

*(The remaining 19 features involve aggregated statistics of the Markov Blankets, e.g., mean MI between $X$ and all neighbors of $Y$.)*
